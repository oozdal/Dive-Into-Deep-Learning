# Dimensionality Reduction

```python
import pandas as pd
import time
import trimap
import umap
import pacmap
from sklearn.manifold import TSNE

# Read data
df = pd.read_csv('Data/Wine.csv')

# Algorithms to test
algorithms = {
    't-SNE': TSNE(),
    'UMAP': umap.UMAP(),
    'TriMAP': trimap.TRIMAP(),
    'PaCMAP': pacmap.PaCMAP()
}

# Create embeddings for each algorithm & save fitting time
results = {}
for name, algorithm in algorithms.items():

    # Calculate embeddings
    t0 = time.time()
    embedding = algorithm.fit_transform(df.values)

    # Store results & runtime
    results[name] = {
        'embedding': pd.DataFrame(embedding, columns=['x', 'y']), 
        'time': time.time() - t0
    }
```

```python
results
```

```python
import matplotlib.pyplot as plt

fig, ((ax1, ax2, ax3, ax4)) = plt.subplots(nrows=1, ncols=4, sharex=False, sharey=True, figsize=(20,20))
ax1.scatter(results['PaCMAP']['embedding'].values[:, 0], results['PaCMAP']['embedding'].values[:, 1], c=df.Customer_Segment) 
ax2.scatter(results['t-SNE']['embedding'].values[:, 0], results['t-SNE']['embedding'].values[:, 1], c=df.Customer_Segment) 
ax3.scatter(results['UMAP']['embedding'].values[:, 0], results['UMAP']['embedding'].values[:, 1], c=df.Customer_Segment) 
ax4.scatter(results['TriMAP']['embedding'].values[:, 0], results['TriMAP']['embedding'].values[:, 1], c=df.Customer_Segment) 

ax1.title.set_text('PaCMAP')
ax2.title.set_text('t-SNE')
ax3.title.set_text('UMAP')
ax4.title.set_text('TriMAP')
```
